```python
import pandas as pd
import numpy as np
df=pd.read_csv("reviews_detail.csv",encoding='utf-8')
```

发现reviews_detail.csv文件中有多个换行符问题，导致后面的计算出现报错，
换行符就是一个excel表中的单元格，里面有多行数据，这样不利于进行数据分析，所以要先进行处理

```python
df_1=df['comments'].replace('\n', '').replace('\r', '').replace('\t', '')
```

```python
# 此处遍历comments，除去换行符，将\n，\r,\t全部除去
from snownlp import SnowNLP

coms=[]
for i in df_1:
    coms.append(str(i).replace('\n', '').replace('\r', '').replace('\t', ''))#除去换行符
```

```python
len(coms)
```

score为各个评论的分数，接下来进行处理，大于0.5的为正向评论，设置为1，其余的为负向评论，设置为0

```python
import numpy as np
a=np.load('score.npy')
score=a.tolist()
```

```python
score_1=[]#为处理后的评论，为0-1向量列表，即列表元素不是1就是0.
for i in score:
    if i>0.5:
        i=1
        score_1.append(i)
    else:
        score_1.append(0)
```

```python
#countX函数用来统计score_1列表中元素为1的个数
def countX(lst, x):
    count = 0
    for ele in lst:
        if (ele == x):
            count = count + 1
    return count
 

x = 1
print(countX(score_1, x))#输出score_1列表中，变量为1的个数，剩下的变量为0
```

```python
len(score_1)
```

```python
score_pos=countX(score_1, x)/len(score_1)#评论评分
```

```python
score_pos
```

```python
score_neg=1-score_pos
score_neg
```

```python
score_2=[]#score_2为我们想要的“评论评分”列的所有值
for i in score_1:
    if i==1:
        score_2.append(score_pos)
    else:
        score_2.append(score_neg)
```

```python
df['评论评分']=score_2#增加“评论评分列”
```

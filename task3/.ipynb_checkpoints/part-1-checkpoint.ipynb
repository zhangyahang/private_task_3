```python
df1=df
```

在数据集增加一列：一类价格的平均值

```python
df1['一类价格的平均值']=df1.groupby('neighbourhood')['price'].transform('mean')
```

在数据集再增加一列：二类价格的平均值

```python
df1['二类价格的平均值']=df1.groupby(['neighbourhood','room_type'])['price'].transform('mean')
```

```python
df1
```

判断是否溢价，并在数据集后面增加“是否一类溢价”列

```python

def function(a,b):
    if a>b:
        return 1
    else:
        return 0
df1['是否一类溢价']=df1.apply(lambda df1:function(df1['price'],df1['一类价格的平均值']),axis=1)

```

```python
df1
```

```python

def function(a,b):
    if a>b:
        return 1
    else:
        return 0
df1['是否二类溢价']=df1.apply(lambda df1:function(df1['price'],df1['二类价格的平均值']),axis=1)
```

```python
df1
```

```python
df1.to_csv('result.csv',index=None)#将df1输出到result.cvs文件，进行保存
```

```python
#取前10个关键词
# 导入包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import jieba.analyse
# withWeight=True一并返回关键词、权重值
keywords = jieba.analyse.extract_tags(
comments, topK=10, withWeight=True, allowPOS=("n", "v")) # 指定名词和动词

# 变成dataframe格式
comm = pd.DataFrame(keywords)
word_list=[]
for item in keywords:
    print(item[0], item[1])
    word_list.append(item[0])

```
